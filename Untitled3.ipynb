```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/healthcare_patient_journey_data.csv'  # Replace with the actual file path
df = pd.read_csv(file_path)


```

```python
# Display basic information about the dataset
print(df.info())

# Display the first few rows of the dataset
print(df.head())

```

```python
# Visualize Waiting Time Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['WaitingTime'], bins=20, color='skyblue', edgecolor='black')
plt.title('Waiting Time Distribution')
plt.xlabel('Waiting Time (days)')
plt.ylabel('Frequency')
plt.show()

# Time Series Analysis
df['AdmissionDate'] = pd.to_datetime(df['AdmissionDate'])
df['DischargeDate'] = pd.to_datetime(df['DischargeDate'])
df['AdmissionDate'].value_counts().sort_index().plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Daily Patient Admissions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Admissions')
plt.show()

# Patient Segmentation
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WaitingTime'], c='coral', alpha=0.5)
plt.title('Patient Age vs. Waiting Time')
plt.xlabel('Age')
plt.ylabel('Waiting Time (days)')
plt.show()

# Root Cause Analysis
plt.figure(figsize=(10, 6))
plt.scatter(df['BottleneckProbability'], df['WaitingTime'], c='green', alpha=0.5)
plt.title('Bottleneck Probability vs. Waiting Time')
plt.xlabel('Bottleneck Probability')
plt.ylabel('Waiting Time (days)')
plt.show()
```
